```python
import numpy as np
import pandas as pd



```

```python
dfa=pd.read_csv('employee_data.csv',encoding='latin-1')
```

```python
# 1. Rename Columns:
#  The column names are notstandardized. Rename the following columns:
# ▪ full_name to name
# ▪ emp_id to employee_id
# ▪ emp_statusto employment_status
d={'full_name':'name','emp_id':'employee_id','emp_status':'employment_status'}
dfa=dfa.rename(columns=d)
dfa
```

```python
# 2. Rearrange Columns:
# Rearrange the columns in this order:
#  ▪employee_id, name, age, gender, department,join_date,
# exp_years,salary,employment_status
L=[0,1,2,3,4,6,7,5,8]
dfa=dfa[dfa.columns[L]]
```

```python
# 3. Update Column Values(Mapping):
# In the gender column, map the values as follows:
#  male and Male should be updated to M
# female and Female should be updated to F
d={'male':'M','female':'F','Male':'M','Female':'F'}
dfa['gender']=dfa['gender'].map(d)
dfa
```

```python
# 4. Remove Unwanted Characters:
#  Remove any leading or trailing spaces in the name column.
dfa['name']=dfa['name'].str.strip()
dfa
```

```python
# 5. Change Data Type:
# o Convert the age column to integer type, and the salary column to float type
avg=dfa['age'].mean()
dfa['age']=dfa['age'].fillna(avg)
dfa['age']=dfa['age'].astype(int)
dfa['salary']=dfa['salary'].astype(float)
dfa
```

```python
# 6. Replace Specific Values:
#  In the employment_status column, replace missing values with 'contract'.
dfa['employment_status']=dfa['employment_status'].fillna('contract')
```

```python
# 7. Handle Missing Values:
# o Fill missing values in the age column with the average age of the employees.
# o Fill missing values in the salary column with the median salary.
# o For missing join_date, use forward fill (ffill).
dfa['age']=dfa['age'].fillna(avg)
ag=dfa['salary'].median()
dfa['salary']=dfa['salary'].fillna(ag)
dfa['join_date']=dfa['join_date'].ffill()
dfa
```

```python
# 8. Remove Duplicates:
# o Ensure there are no duplicate rows in the dataset.
dfa=dfa.drop_duplicates()
```

```python
# 9. Apply String Function:
# o Capitalize the first letter of each word in the department column.
dfa['department']=dfa['department'].str.title()
dfa
```

```python
# 10. Drop Unwanted Columns:
# o Remove the exp_years column from the dataset.
dfa=dfa.drop('exp_years',axis=1)
```

```python
# 11. Remove Specific Rows:
# o Drop rows where salary is above 80,000 to focus on mid-level employees
dfa=dfa.drop(dfa[dfa['salary']>80000].index)
dfa

```

```python
# 12. Reset Index:
#  After cleaning, reset the index of the DataFrame.
dfa=dfa.reset_index(drop=True)
dfa
```

```python
# 13. Set Column as Index:
# o Set the employee_id as the index of the DataFrame.
dfa=dfa.set_index('employee_id')
```
