```python
import numpy as np
import pandas as pd
```

```python
dft=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/titanic.csv',encoding='latin-1')
dft
```

```python
#1)write a condition to fetch all passangers whose age>30
dft[['name']][dft['age']>30]
```

```python
#2)write a condition to fetch all passangers whose pclass==3
dft[['name','pclass']][dft['pclass']==3]
```

```python
#3)write a condition to fetch all femail passangers who survived
dft[['name','survived','sex']][(dft['sex']=='female') & (dft['survived']==1)]
```

```python
#4 write a condition to fetch all passangers who are male OR who paid fare>100
dft[['name','sex','fare']][(dft['sex']=='male') | (dft['fare']>100)]
```

```python
#5)write a condition to fetch all passangers whose age is not null
dft[['name','age']][dft['age'].notnull()]
```

```python
#6)write a condition to fetch all passangers whose age is between 20 and 40
dft[['name','age']][(dft['age']>20) & (dft['age']<40)]
```

```python
#7)write a condition to fetch all passangers who survived AND paid fare greater then the average Fare
dft[['name','survived','fare']][(dft['survived']==1) & (dft['fare'] > dft['fare'].mean())]
```

```python
#8 write the condition to fetch all passangers who did not survive AND were male AND paid Fare<20
dft[['name','survived','sex']][(dft['survived']==0) & (dft['sex']=='male') & (dft['fare']<20)]
```

```python
#9 write a condition to fetch passangers who have both siblings (SibSp>0) AND parents?children (parch>0)
dft[['name','sibsp','parch']][(dft['sibsp']>0) & (dft['parch']>0)]
```
